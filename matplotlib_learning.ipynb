```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.random.randn(100)
nut_butter_shop = { "ALMOND_BUTTER" : 10,
                   "PEANUT_BUTTER" : 8,
                   "CASHEW_BUTTER" :12
}
butters = nut_butter_shop.keys()
prices = nut_butter_shop.values()

mycolors = np.array(["RED","GREEN","YELLOW"])

fig, ((ax1,ax2),(ax3,ax4),(ax5,ax6)) = plt.subplots(nrows = 3,
                                      ncols=2,
                                      figsize=(20,20))

ax1.plot(butters,prices, linestyle="dashed", marker= 'o', c='#4CAF50'); #LINE
ax1.grid()
ax1.set_title("LINE_PLOT")

ax2.scatter(butters,prices, c = mycolors) # SCATTER PLOT
ax2.grid()
ax2.set_title("SCATTER_PLOT")

ax3.bar(butters,prices, width=0.15, color=mycolors)#BAR PLOT
ax3.set_title("BAR_PLOT")

ax4.hist(np.random.randn(500));# HISTOGRAM
ax4.set_title("HISTO_PLOT")

ax5.barh(butters,prices,height=0.25, color=mycolors)#HORIZONTAL BAR
ax5.set_title("HORIZONTAL_PLOT")

ax6.pie(prices, labels=butters, startangle=90)#PIECHART
ax6.legend(title="BUTTER_RATES");#index to piechart
ax6.set_title("PIECHART_PLOT");
```
